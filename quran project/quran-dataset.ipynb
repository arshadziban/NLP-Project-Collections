```python
import pandas as pd
import spacy
```

```python
# Load NLP Model
nlp = spacy.load("en_core_web_sm")
```

```python
# Load dataset
df = pd.read_csv("/kaggle/input/quran-english-translation-tafaseer/main_df.csv")
```

```python
df
```

```python
def preprocess(text):
    if pd.isna(text): 
        return ""
    doc = nlp(text.lower())  
    words = [token.lemma_ for token in doc if not token.is_stop]  
    return " ".join(words)
```

```python
df["Processed_Text"] = df["Translation1"].apply(preprocess)
```

```python
df
```

```python
df.to_csv("quran_dataset_cleaned.csv", index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load Preprocessed Data
df2 = pd.read_csv("quran_dataset_cleaned.csv")
```

```python
df2
```

```python
# Convert verses into numerical format
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["Processed_Text"].fillna(""))
```

```python
def find_best_match(query):
    query_vec = vectorizer.transform([query])  # Convert query to vector
    similarities = cosine_similarity(query_vec, X)  # Compare with all verses
    best_match_idx = similarities.argmax()  # Get best matching verse index
    return df.iloc[best_match_idx]
```

```python
if __name__ == "__main__":
    query = input("Enter a keyword or phrase: ")
    result = find_best_match(query)

    print("\nğŸ“– Best Matched Verse:")
    print(f"ğŸ“Œ Surah {result['Surah']}, Ayah {result['Ayat']}")
    print(f"ğŸ“œ Arabic: {result['Arabic']}")
    print(f"ğŸ“ Translation: {result['Translation1']}")
    print(f"ğŸ“– Tafseer: {result['Tafaseer1']}")
```
